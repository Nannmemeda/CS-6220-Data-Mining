## 1 Online Retail Market Basket Analysis [50 points]

```python
import pandas as pd
pd.set_option("max_colwidth", 150)
f = "https://github.com/cs6220/cs6220.spring2019/raw/master/data/Online%20Retail.xlsx"
df = pd.read_excel(f)
basket = (df[df["Country"] == "United Kingdom"]
        .groupby(["InvoiceNo", "Description"])["Quantity"]
        .sum().unstack().reset_index().fillna(0)
        .set_index("InvoiceNo")) # transform transactions into baskets of items
basket_sets = basket.applymap(lambda x: 1 if x >=1 else 0) # convert counts to booleans
```

### 1.1 Frequent Itemset Generation

Use the Apriori algorithm to generate frequent itemsets with up to 2 items per set. Explore dif-
ferent support thresholds until you obtain at least 5 results for both 1- and 2-itemsets. Note that
mlxtend offers an implementation of the Apriori algorithm (see mlxtend.frequent patterns).

#### [10 pts] What are the top 5 1-itemsets with the highest support?

```python
# Generate the frequent itemset and find the top 5 1-itemset support

frequent_itemsets = apriori(basket_sets, min_support = 0.05, use_colnames = True)
frequent_itemsets = frequent_itemsets.sort_values("support", ascending = False)
top_5_1item = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 1][:5]
top_5_1item
```

#### [10 pts] What are the top 5 2-itemsets with the highest support?

```python
# Find the top 5 2 itemsets with the highest support

frequent_itemsets = apriori(basket_sets, min_support=0.02, use_colnames=True)
frequent_itemsets = frequent_itemsets.sort_values(by="support", ascending=False)
top_5_2items = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 2][:5]
top_5_2items
```

#### [5 pts] What is the highest support value for the 1-itemsets?

```python
# Find the highest support value for 1-itemsets

frequent_itemsets = apriori(basket_sets, min_support = 0.05, use_colnames = True)
frequent_itemsets = frequent_itemsets.sort_values("support", ascending = False)
highest_support_1itemset = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 1]['support'].values[0]
print("The highest support value for 1-itemsets is: ",highest_support_1itemset)
```

#### [5 pts] What is the highest support value for the 2-itemsets?

```python
# Find the highest support value for 2-itemsets

frequent_itemsets = apriori(basket_sets, min_support = 0.02, use_colnames = True)
frequent_itemsets = frequent_itemsets.sort_values("support", ascending = False)
highest_support_2itemset = frequent_itemsets[frequent_itemsets['itemsets'].apply(len) == 2]['support'].values[0]
print("The highest support value for 2-itemsets is: ",highest_support_2itemset)
```

I have generated the rules with top 5 highest lift. Lift is calculated as the ratio of the observed support for  given itemset to the expected support if the items were independent. A lift value greater than 1 indicates that the items are more likely to appear together in transactions than if they were independent.

The first rule is (salary > 50K) => (marital status is Married-civ-spouse), and it has 1.855609 lift. This means when people with annual income greater than $50,000 have approximately 1.86 times more likely to be married  and have a civil spouse, compared to their income and marital status were independent.

This means when people have a civil spouse, and his/her race is white, will have approximately 1.34 times more likely to be a male, compared to their race, sex and marital status were independent.

The forth rule is (native-country_ United-States, marital-status_ Married-civ-spouse) => (native-country_ United-States, sex_ Male), and it has 1.332888 lift. #### [10 pts] Compare the top rules using the two interestingness measures for the same levels of support (use at least two different levels of support) and comment on your findings.

```python
# Case when min_support is 0.1

frequent_itemsets_1 = apriori(df_adult, min_support = 0.1, use_colnames = True)
rule_1_confidence = association_rules(frequent_itemsets_1, metric = 'confidence', min_threshold = 0.7)
rule_1_lift = association_rules(frequent_itemsets_1, metric = 'lift', min_threshold = 1.2)
```

```python
print("Confidence:")
rule_1_confidence.sort_values(by='confidence', ascending=False)[:3]
```

```python
print("Lift:")
rule_1_lift.sort_values(by='lift', ascending=False)[:3]
```

Compared the same levels support with different measures(confidence and lift) and same level support, we can find that it will generate different top rules when we use different measures. 

For example, the top 1 rule using confidence measure is (race_ White, occupation_ Craft-repair) => (sex_ Male), and its confidence is 0.953979, lift is 1.425540. These means we have approximately 95.40% probability that the is a male when the person's race is white and occupation is craft-repair. Also, when people race is white and occupation is craft repair, it has approximately 1.43 times more likely to be a male.

However, the top 1 rule using lift measure is (race_ White, relationship_ Own-child) => (marital-status_ Never-married, native-country_ United-States), and its confidence is 0.842068, lift is 2.862364. Thease means we have approximately 84.21% probability that the person's marital status is never married and his/her native country is United States when the person's race is white and relationship is own-child. Also, when people race is white and relationship is own-child, it has approximately 2.86 times more likely that their marital status is never married and his/her native country is United States.

Therefore, the different measures will generate different top rules. We should choose the measures by needs.

```python
# Case when min_support is 0.05

frequent_itemsets_2 = apriori(df_adult, min_support = 0.05, use_colnames = True)
rule_2_confidence = association_rules(frequent_itemsets_2, metric = 'confidence', min_threshold = 0.7)
rule_2_lift = association_rules(frequent_itemsets_2, metric = 'lift', min_threshold = 1.2)
```

```python
print("Lift:")
rule_2_lift.sort_values(by='lift', ascending=False)[:3]
```

With different min_support value, we can see that the transactions and the top itemsets has totally changed. For example, when min_support is 0.1, under confidence measure, the top rule is (race_ White, occupation_ Craft-repair) => (sex_ Male). However, when min_support is 0.05, the top rule is (race_ White, marital-status_ Married-civ-spouse, occupation_ Craft-repair, native-country_ United-States) => (sex_ Male). And their confidence or lift or other measures are increased as well by the min_support decreasing. For example, under confidence measure and min_support is 0.1, the top rule's confidence is 0.953979, while it becomes 0.991344 when the the min_support decrease to 0.05.

Therefore, different min_support value will generate different quantity itemsets, and may influence the itemsets that we choose in the final decision. The purpose of setting a minimum support threshold is to reduce the number of candidate itemsets that need to be considered for rule generation, as well as to control the size of the rules that are generated. If the minimum support threshold is set too low, then many meaningless or uninteresting itemsets will be generated. If the threshold is set too high, then very few itemsets will be generated and many interesting patterns may be missed.
