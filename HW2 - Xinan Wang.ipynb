### 2.3 Feature Selection

#### •[10 pts] To get an idea of their relative importance, estimate the mutual information between the numeric features and the class column, ‘SalePrice’. Note that scikit-learn may be used to estimate mutual information (see sklearn.feature selection).

```python
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import mutual_info_classif
from sklearn.feature_selection import SelectKBest
from sklearn.datasets import make_swiss_roll

# Select the numeric features
numeric_df = df.select_dtypes(include=[np.number]).columns
numeric_df = numeric_df.drop(['SalePrice'])
X = df[numeric_df]
y = df['SalePrice']

mutual = mutual_info_regression(X, y)

print("Mutual information of each feature:")
for feature, mi in zip(numeric_df, mutual):
    print("{}: {}".format(feature, mi))
```

#### •[10 pts] What are the top 5 numeric features ranked by mutual information? Note that features with a higher estimated mutual information are considered more informative.

```python
# Select the k best features.
selector = SelectKBest(score_func=mutual_info_regression, k=5)
selector.fit(X, y)

# Get indices of columns to keep.
idxs_selected = selector.get_support(indices=True)

# Create new DataFrame with only desired columns.
X_new = X.iloc[:, idxs_selected]

# Show the top 5 features.
print(X_new.head())

X_selected = X_new.columns
X_selected = df[X_selected]
```

#### [5 pts] How do you expect the values for the top-ranked feature to affect the sales price (i.e., would you expect the sales price to increase when its values go up or down)? Why?

```python
# Use Logistic Regression to find out the positive or negative effect on a feature

import shap
import xgboost as xgb

model = xgb.XGBRegressor(n_estimator=500, max_depth=20,learning_rate=0.1,subsample=0.8,random_state=33)
model.fit(X_selected,y)

explainer = shap.TreeExplainer(model)
shap_values = explainer(X_selected)
shap.summary_plot(shap_values,X_selected)
```

According to the feature value plot, the 'Overall Qual' has the most largest influence on SalePrice. 'Gr Liv Area', '1st Flr SF', 'Year Built' take the second, third, forth place, and 'MS SubClass' takes the final place.

'Overall Qual' means the rates the overall material and finish of the house. As we can see from the plot, the red scatters of 'Overall Qual' are mostly located on the right side of the y axis, which means as the sale price increasing, we mostly expect the 'Overall Qual' go up.

'Gr Liv Area' means the above ground living area square feet. As we can see from the plot, the red scatters of 'Gr Liv Area' are mostly located on the right side of the y axis, which means as the sale price increasing, we mostly expect the 'Gr Liv Area' go up.

'1st Flr SF' means the first floor square feet. As we can see from the plot, the red scatters of '1st Flr SF' are mostly located on the right side of the y axis, which means as the sale price increasing, we mostly expect the '1st Flr SF' go up.

'Year Built' means the year the house is built. As we can see from the plot, the red scatters of 'Year Built' are mostly located on the right side of the y axis, which means as the sale price increasing, we mostly expect the 'Year Built' go up.

'MS SubClass' identifies the type of dwelling involved in the sale. As we can see from the plot, the red scatters of 'MS SubClass' are mostly located on the left side of the y axis, which means as the sale price increasing, we mostly expect the 'Gr Liv Area' go down.
